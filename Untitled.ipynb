```python
import pandas as pd
import numpy as np
```

```python
# Reading data
data=pd.read_csv('data.csv')
data=np.array(data.iloc[:, :])
```

```python
# Split into training, validation and testing parts
from sklearn.model_selection import train_test_split
train_size=0.7
val_size=0.5 # Decide between validation and testing
X_train, X_tem = train_test_split(data, test_size=1-train_size, shuffle=False)
X_val, X_test = train_test_split(X_tem, test_size=1-val_size, shuffle=False)
```

```python
import torch
from torch.autograd import Variable
import torch.nn as nn
from sklearn.preprocessing import MinMaxScaler
```

```python
def prepare_data(X, window_size, sc=None):
    '''
    window_size : number of timesteps to look back for predicting the next timestep
    '''
    if sc is None:
        sc = MinMaxScaler()
        X = sc.fit_transform(X)
    else:
        X=sc.transform(X)
    
    x=[]
    y=[]
    dim=X.shape[1]
    for i in range(window_size, X.shape[0]):
        x.append(X[(i-window_size):i, :].reshape(window_size, dim))
        y.append(X[i, :].reshape(1, dim))
    
    x=np.array(x); y=np.squeeze(np.array(y));
    
    return Variable(torch.Tensor(x)), Variable(torch.Tensor(y)), sc
```

```python
class LSTM(nn.Module):

    def __init__(self, num_classes, input_size, hidden_size, num_layers):
        super(LSTM, self).__init__()
        
        self.num_classes = num_classes
        self.num_layers = num_layers
        self.input_size = input_size
        self.hidden_size = hidden_size
        
        self.lstm = nn.LSTM(input_size=input_size, hidden_size=hidden_size,
                            num_layers=num_layers, batch_first=True)
        
        self.fc = nn.Linear(hidden_size, num_classes)

    def forward(self, x):
        h_0 = Variable(torch.zeros(
            self.num_layers, x.size(0), self.hidden_size))
        
        c_0 = Variable(torch.zeros(
            self.num_layers, x.size(0), self.hidden_size))
        
        # Propagate input through LSTM
        ula, (h_out, _) = self.lstm(x, (h_0, c_0))
        
        h_out = h_out.view(-1, self.hidden_size)
        
        out = self.fc(h_out)
        
        return out
```

```python
def predict(x, sc, model):
    model=model.eval()
    outputs = model(x)
    outputs=sc.inverse_transform(model(x).data.numpy())
    return outputs

def predict_rolling(x, sc, model, num_timesteps):
    model=model.eval()
    outputs = None
    # x shape : (1, window_size, num_series)
    for i in range(num_timesteps):
        out=model(x).reshape(1, 1, x.shape[-1])
        if outputs is None:
            outputs=out
        else:
            outputs=torch.cat([outputs, out], axis=0)
        x=out
    
    outputs=outputs.reshape(num_timesteps, x.shape[-1])
    outputs=sc.inverse_transform(outputs.data.numpy())
    return outputs
```

```python
import random
RANDOM_SEED = 42
np.random.seed(RANDOM_SEED)
torch.manual_seed(RANDOM_SEED)
random.seed(RANDOM_SEED)
torch.cuda.manual_seed_all(RANDOM_SEED)

# Prepare data #
window_size=5
# Append last `window_size` timesteps to val/test
X_val=np.vstack([X_train[X_train.shape[0]-window_size:X_train.shape[0], :], X_val])
X_test=np.vstack([X_val[X_val.shape[0]-window_size:X_val.shape[0], :], X_test])
x_train,y_train,sc=prepare_data(X_train, window_size)
x_val,y_val,sc=prepare_data(X_val, window_size, sc)
x_test,y_test,sc=prepare_data(X_test, 5, sc)

# Train #
num_epochs = 1000
learning_rate = 0.01

input_size = 30
hidden_size = 200
num_layers = 1

num_classes = 30

lstm = LSTM(num_classes, input_size, hidden_size, num_layers)
model_final=None
min_val_loss=1e20

criterion = torch.nn.MSELoss()    # mean-squared error for regression
optimizer = torch.optim.Adam(lstm.parameters(), lr=learning_rate)
#optimizer = torch.optim.SGD(lstm.parameters(), lr=learning_rate)

# Train the model
for epoch in range(num_epochs):
    outputs = lstm(x_train)
    optimizer.zero_grad()
    
    # obtain the loss function
    loss = criterion(outputs, y_train)
    
    loss.backward()
    
    optimizer.step()
    if epoch % 100 == 0:
        lstm=lstm.eval()
        val_loss = criterion(lstm(x_val), y_val)
        # Make validation prediction #
#         x_val_start=x_train[-1, :, :].reshape(1, window_size, x_train.shape[-1])
#         val_preds=predict_rolling(x_val_start, sc, lstm, y_val.shape[0])
#         val_loss=np.mean((sc.transform(val_preds)-y_val.data.numpy())**2)
        
        if val_loss <= min_val_loss:
            min_val_loss=val_loss
            model_final=lstm
        print("Epoch: %d, train loss: %1.5f, val loss: %1.5f" % (epoch, loss.item(), val_loss.item()))
        lstm=lstm.train()
```

```python
import matplotlib.pyplot as plt

def make_grid_plot(x,y,lstm):
    fig, axs = plt.subplots(5, 6)
    preds=predict(x, sc, lstm)
    for i in range(5):
        for j in range(6):
            idx=i*5+j
            axs[i,j].plot(preds[:, idx])
            axs[i,j].plot(sc.inverse_transform(y)[:, idx])
    return fig
            
def make_grid_plot_rolling_val(y,lstm):
    fig, axs = plt.subplots(5, 6)
    x_val_start=x_train[-1, :, :].reshape(1, window_size, x_train.shape[-1])
    preds=predict_rolling(x_val_start, sc, lstm, y_val.shape[0])
    
    for i in range(5):
        for j in range(6):
            idx=i*5+j
            axs[i,j].plot(preds[:, idx])
            axs[i,j].plot(sc.inverse_transform(y)[:, idx])
```
